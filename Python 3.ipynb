```python
num = int(input("Enter a number: "))
factorial = 1
i = 1

while i <= num:
    factorial = factorial * i
    i = i + 1

print(factorial)
```

5. Take an integer as input and reverse its digits (e.g. 123 -> 321).


```python
num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reversed number is:", rev)
```

55. Write a function that takes a list of dictionaries (students with marks) and returns the
topperâ€™s name. 

```python
def find_topper(students):
    if not students:
        return None  # handle empty list
    topper = students[0]
    for student in students:
        if student["marks"] > topper["marks"]:
            topper = student
    return topper["name"]
students = [
    {"name": "Ankita", "marks": 85},
    {"name": "Sneha", "marks": 92},
    {"name": "Priya", "marks": 88}]
topper_name = find_topper(students)
print("Topper:", topper_name)
```
