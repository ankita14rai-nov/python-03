```python
# Create a dictionary of students and their marks
students = {
    "Alice": 85,
    "Bob": 78,
    "Charlie": 92,
    "David": 88,
    "Eva": 90
}

# Print all student names using keys()
print(students.keys())
print(students.values())
print(students.items())
print(students.get("Rahul"))

students.update({
    "David": 88,
    "Eva": 90,
    "Frank": 75
})

print(students)
removed_mark = students.pop("David", "Not Found")

# Print what was returned
print(removed_mark)

# Print the dictionary after pop
print(students)
students.clear()
```

Create  a dict make a copy using .copy() and update the copy without affecting the original.

```python
# Original dictionary
students = {
    "Alice": 85,
    "Bob": 78,
    "Charlie": 92
}

# Make a copy of the dictionary
students_copy = students.copy()

# Update the copy
students_copy["David"] = 88
students_copy["Alice"] = 95  # changing a value in the copy

# Print both dictionaries
print("Original dictionary:", students)
print("Updated copy:", students_copy)
```

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for v in vowels:
        count += len(text.split(v)) - 1
    return count
sample_text = "Hello, welcome to Python programming."
vowel_count = count_vowels(sample_text)
print("Number of vowels:", vowel_count)
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
num = 10
fact = factorial(num)
print("Factorial:", fact)
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
num = 5
fact = factorial(num)   
print("Factorial:", fact)
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

def print_fibonacci(n):
    for i in range(n):
        print(fibonacci(i), end=" ")
print_fibonacci(10)
```

```python
def find_topper(students):
    if not students:
        return None  # handle empty list
    topper = students[0]
    for student in students:
        if student["marks"] > topper["marks"]:
            topper = student
    return topper["name"]
students = [
    {"name": "Ankita", "marks": 85},
    {"name": "Sneha", "marks": 92},
    {"name": "Priya", "marks": 88}]
topper_name = find_topper(students)
print("Topper:", topper_name)
```
